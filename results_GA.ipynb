```python
import tsplib95
import GA
import pandas as pd
from matplotlib import pyplot as plt

# https://people.sc.fsu.edu/~jburkardt/datasets/tsp/five_d.txt
# is a set of 5 cities. The minimal tour has length 19.

# Read the distances matrix from a txt file
df = pd.read_csv('TSP files/five_d.txt' , delimiter='\s+', header=None, dtype=int)
print(df.values)

cities_GA1 = list(range(5))
print("cities : " , cities_GA1)

# Convert the matrix to a list of lists
distances1 = df.values.tolist()
print(distances1)

num_generations = 30
GA_1 = GA.GeneticAlgorithm(cities_GA1, distances1, 0, 16, num_generations)
max_index = GA_1.run()

true_minimal_distnce = 19
plt.plot([true_minimal_distnce] * num_generations,'r-')
plt.plot(GA_1.minimal_distance_progress)
plt.xlabel("genertion")
plt.ylabel("minimal distance(cost)")
plt.title("evolution of the minimum total traveling distance with generation")
plt.legend()
```
